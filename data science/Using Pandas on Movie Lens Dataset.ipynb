```python
# pass in column names for each CSV
u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
users = pd.read_csv(DATA_DIR + 'ml-100k/u.user', sep='|', names=u_cols)

r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv(DATA_DIR + 'ml-100k/u.data', sep='\t', names=r_cols)

# the movies file contains columns indicating the movie's genres
# let's only load the first five columns of the file with usecols
m_cols = ['movie_id', 'title', 'release_date', 'video_release_date', 'imdb_url']
movies = pd.read_csv(DATA_DIR + 'ml-100k/u.item', sep='|', names=m_cols, usecols=range(5))
```

```python
chicago_headers = ['name', 'title', 'department', 'salary']
chicago = pd.read_csv(DATA_DIR + 'city-of-chicago-salaries.csv', header=False, names=chicago_headers,
                      converters={'salary': lambda x: float(x.replace('$', ''))})
print chicago.head()
```

```python
movie_ratings = pd.merge(movies, ratings)
lens = pd.merge(movie_ratings, users)
```

```python
by_title = lens.groupby('title')
most_rated = by_title.size().order(ascending=False)
most_rated.head()
```

```python
movie_states = by_title.agg({'rating': [np.size, np.mean]})
print movie_states.sort([('rating', 'size')], ascending=False).head()
```

```python
atleast100 = movie_states[('rating', 'size')] >= 100
# print atleast100
print movie_states[atleast100].sort([('rating', 'mean')], ascending=False).head()
```

```python
lens.to_sql('movie_lens_100k', DB_ENGINE, if_exists='replace', index=False)
```

```python
most50 = lens.groupby('movie_id').size().order(ascending=False)[:50]
```

```python
users.age.hist(bins=50)
```

```python
labels = [str(i)+'x' for i in range(8)]
lens['age_group'] = pd.cut(lens.age, range(0,81,10), right=False, labels=labels)
print lens[['age', 'age_group']].drop_duplicates()[:10]
```

```python
lens.groupby('age_group').agg({'rating':[np.size, np.mean]})
```

```python
lens.set_index('movie_id', inplace=True)
by_age = lens.ix[most50.index].groupby(['title', 'age_group'])
by_age.rating.mean().head(20)
```

```python
by_age.rating.mean().unstack(1).fillna(0)
```

```python
lens.reset_index('movie_id', inplace=True)
```

```python
pivoted = lens.pivot_table(rows=['movie_id', 'title'],
                           columns=['sex'],
                           values='rating',
                           fill_value=0)
print pivoted.head()
```

```python
pivoted['diff'] = pivoted.F - pivoted.M
print pivoted.head()
```

```python
pivoted.reset_index('movie_id', inplace=True)
```

```python
disagreements = pivoted[pivoted.movie_id.isin(most50.index)]['diff']
disagreements.order().plot(kind='barh', figsize=[9,15])
plt.title('Male vs Female Avg. Rating\n(Diff > 0 = Favored by Men)')
```
